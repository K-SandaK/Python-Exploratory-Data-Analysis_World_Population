The first visualization will be a **Heatmap** to show correlation 

```python
numeric_df= df.select_dtypes(include=['float64', 'int64'])
correlation= numeric_df.corr()
correlation
```

```python
sns.heatmap(numeric_df.corr(), annot =True)
plt.rcParams['figure.figsize'] = (12,8)
plt.show
```

Group by Continent and show the growth rate 

```python
df.select_dtypes(include='number').groupby(df['Continent'])[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean()
```

```python
df2=df.select_dtypes(include='number').groupby(df['Continent'])[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean()
df2
```
